```python
from pydantic import BaseModel
```

```python
class Person(BaseModel):
    first_name: str
    last_name: str
    age: int
```

```python
p = Person(first_name="John", last_name="Doe", age=30)
```

```python
p
```

```python
from pydantic import ValidationError
```

```python
try:
    Person()
except ValidationError as e:
    print(e)
```

```python
try:
    Person()
except ValidationError as e:
    print(e.json())
```

```python
class Person(BaseModel):
    first_name: str
    last_name: str
    age: int | None = None
```

```python
try:
    Person(first_name="John", last_name="Doe")
except ValidationError as e:
    print(e.json())
```
