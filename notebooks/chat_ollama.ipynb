````python
from langchain_ollama import ChatOllama
chat_model = ChatOllama(model="llama3", temperature=0.0)
chat_model
````

````python
template_string = """Translate the text \
that is delimited by triple backticks \
into a style that is {style}. \
text: ```{text}```
"""
````

````python
from langchain_core.prompts import ChatPromptTemplate
prompt_template = ChatPromptTemplate.from_template(template_string)
````

````python
prompt_template.messages[0].prompt
````

````python
prompt_template.messages[0].prompt.input_variables
````

````python
customer_style = """American English \
in a calm and respectful tone
"""
````

````python
customer_email = """
Arrr, I be fuming that me blender lid \
flew off and splattered me kitchen walls \
with smoothie! And to make matters worse, \
the warranty don't cover the cost of \
cleaning up me kitchen. I need yer help \
right now, matey!
"""
````

````python
customer_messages = prompt_template.format_messages(
                    style=customer_style,
                    text=customer_email)
````

````python
print(type(customer_messages))
print(type(customer_messages[0]))
````

````python
print(customer_messages[0])
````

````python
# Call the LLM to translate to the style of the customer message
customer_response = chat_model.invoke(customer_messages)
print(customer_response.content)
````

````python
print(customer_response.response_metadata)
````

````python
if customer_response.usage_metadata:
    print("--- Token Usage ---")
    print(f"Input Tokens: {customer_response.usage_metadata.get('input_tokens')}")
    print(f"Output Tokens: {customer_response.usage_metadata.get('output_tokens')}")
    print(f"Total Tokens: {customer_response.usage_metadata.get('total_tokens')}")
````

````python
def nano_seconds_to_milliseconds(nanoseconds):
    return nanoseconds / 1_000_000

def utc_timestamp_to_local_datetime(utc_timestamp):
    from datetime import datetime, timezone
    dt = datetime.fromisoformat(utc_timestamp.replace('Z', '+00:00'))
    return dt.astimezone()
````

````python
if customer_response.usage_metadata:
    print("--- Token Usage ---")
    print(f"Input Tokens: {customer_response.usage_metadata.get('input_tokens')}")
    print(f"Output Tokens: {customer_response.usage_metadata.get('output_tokens')}")
    print(f"Total Tokens: {customer_response.usage_metadata.get('total_tokens')}")
    print(f"Model: {customer_response.response_metadata.get('model')}")
    print(f"Timestamp: {utc_timestamp_to_local_datetime(customer_response.response_metadata.get('created_at'))}")
    print("--- Duration ---")
    print(f"Total Duration (ms): {nano_seconds_to_milliseconds(customer_response.response_metadata.get('total_duration'))}")
    print(f"Load Duration (ms): {nano_seconds_to_milliseconds(customer_response.response_metadata.get('load_duration'))}")
    print(f"Prompt Eval Duration (ms): {nano_seconds_to_milliseconds(customer_response.response_metadata.get('prompt_eval_duration'))}")
    print(f"Eval Duration (ms): {nano_seconds_to_milliseconds(customer_response.response_metadata.get('eval_duration'))}")
````

## Output Parsers
Lets define how we would like the LLM output to look like.

````python
{"gift": False,
 "delivery_days": 3,
 "price_value": "pretty cheap"
 }
````

````python
customer_review = """\
This leaf blower is pretty amazing.  It has four settings:\
candle blower, gentle breeze, windy city, and tornado. \
It arrived in two days, just in time for my wife's \
anniversary present. \
I think my wife liked it so much she was speechless. \
So far I've been the only one using it, and I've been \
using it every other morning to clear the leaves on our lawn. \
It's slightly more expensive than the other leaf blowers \
out there, but I think it's worth it for the extra features.
"""
````

````python
review_template = """\
For the following text, extract the following information:

gift: Was the item purchased as a gift for someone else? \
Answer True if yes, False if not or unknown.

delivery_days: How many days did it take for the product \
to arrive? If this information is not found, output -1.

price_value: Extract any sentences about the value or price,\
and output them as a comma separated Python list.

Format the output as JSON with the following keys:
gift
delivery_days
price_value

text: {text}
"""
````

````python
from langchain_core.prompts import ChatPromptTemplate

prompt_template = ChatPromptTemplate.from_template(review_template)
print(prompt_template)
````

````python
messages = prompt_template.format_messages(text=customer_review)
messages
````

````python
response = chat_model.invoke(messages)
print(response.content)
````

````python
if response.usage_metadata:
    print("--- Token Usage ---")
    print(f"Input Tokens: {response.usage_metadata.get('input_tokens')}")
    print(f"Output Tokens: {response.usage_metadata.get('output_tokens')}")
    print(f"Total Tokens: {response.usage_metadata.get('total_tokens')}")
    print(f"Model: {response.response_metadata.get('model')}")
    print(f"Timestamp: {utc_timestamp_to_local_datetime(response.response_metadata.get('created_at'))}")
    print("--- Duration ---")
    print(f"Total Duration (ms): {nano_seconds_to_milliseconds(response.response_metadata.get('total_duration'))}")
    print(f"Load Duration (ms): {nano_seconds_to_milliseconds(response.response_metadata.get('load_duration'))}")
    print(f"Prompt Eval Duration (ms): {nano_seconds_to_milliseconds(response.response_metadata.get('prompt_eval_duration'))}")
    print(f"Eval Duration (ms): {nano_seconds_to_milliseconds(response.response_metadata.get('eval_duration'))}")
````
